```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import hvplot.pandas
import plotly.express as px
```

```python
data = pd.read_excel("Lab_3-2_Lending_Club_Transform.xlsx")
data.head()
```

```python
df = data[["loan_amnt","int_rate","dti"]]
df
```

```python
df2 = df.groupby("int_rate").mean("loan_amnt","dti").reset_index()
```

```python
fig = sns.scatterplot(data=df2,y="loan_amnt",x="dti",hue="int_rate")
plt.show()
```

```python
fig2 = df2.plot.scatter(x="loan_amnt",y="dti",c="int_rate")
```

```python
fig3 = df2.hvplot.scatter(y="loan_amnt",x="dti",by = "int_rate")
fig3
```

```python
fig4 = px.scatter(df2, y="loan_amnt",x="dti",color="int_rate")
fig4
```
